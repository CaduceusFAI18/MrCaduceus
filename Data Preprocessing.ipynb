```python
# Load Raw data
df = []
filename_list = []
for file in listdir("data/2014-raw/"):
    filename = "data/2014-raw/" + file
    filename_list.append(filename)
    print("Reading %s " % filename)
    df.append(pd.read_sas(filename))
```

```python
#Only keep data with SEQN, and remove duplicates
df_new = []
for i in range(len(df)):
    if "SEQN" in df[i].columns.values:
        if len(df[i].columns) < 100:
            df_new.append(df[i].drop_duplicates(subset="SEQN"))
    else:
        print('Dataframe %s missing SEQN or too long' % filename_list[i])
```

```python
df1 = reduce(lambda left,right: pd.merge(left, right, on='SEQN', how='outer'), df_new[0:20])
df2 = reduce(lambda left,right: pd.merge(left, right, on='SEQN', how='outer'), df_new[21:40])
df3 = reduce(lambda left,right: pd.merge(left, right, on='SEQN', how='outer'), df_new[41:60])
df4 = reduce(lambda left,right: pd.merge(left, right, on='SEQN', how='outer'), df_new[61:80])
df5 = reduce(lambda left,right: pd.merge(left, right, on='SEQN', how='outer'), df_new[81:100])
df6 = reduce(lambda left,right: pd.merge(left, right, on='SEQN', how='outer'), df_new[101:])
```

```python
df_merge = [df1, df2, df3, df4, df5, df6]
for i in df_merge:
    print("Rows: %s | Cols: %s" %(len(i), len(i.columns)))
```

```python
df1_df2 = pd.merge(df1, df2, on='SEQN', how='outer')
```

```python
df3_df4 = pd.merge(df3, df4, on='SEQN', how='outer')
```
