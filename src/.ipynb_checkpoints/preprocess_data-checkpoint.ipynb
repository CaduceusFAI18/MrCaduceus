```python
import pandas as pd
import numpy as np
import pickle
```

```python
input_files_name  = ['demographic', 'examination', 'medications', 'diet', 'labs', 'questionnaire']
```

```python
# Use engine python to avoide UnicodeDecode error
input_data = {}

for file_name in input_files_name:
    print(file_name)
    input_data[file_name] = pd.read_csv('../data/raw/2014-kaggle/' + file_name + '.csv', encoding='latin-1')
```

```python
# For the questionnarie data, change na to 0
input_data['questionnaire'].fillna(0, inplace=True)
# Preprocessing: 
#     Remove columns which has more than 35% is null
for file_name in input_files_name:
    table = input_data[file_name]
    table = table.replace(r'^\s+$', np.nan, regex=True)
    columns = {}
    columns_to_be_removed = {}
    
    print(file_name, table.shape)
#     table = table._get_numeric_data()
    for column_name in table:
        column_n_na = table[column_name].isna().sum()
        columns[column_name] = column_n_na/table.shape[0]
        if (column_n_na / table.shape[0]) > 0.35: 
            columns_to_be_removed[column_name] = column_n_na / table.shape[0]
        
    input_data[file_name] = table.drop(columns=list(columns_to_be_removed.keys()))
    print('Columns and null ration')
    print(str(columns))
    print('Number of columns removed: ', len(columns_to_be_removed))
    print('Columns removed and null ratio')
    print(str(columns_to_be_removed))
    print('#############')
```

```python
# Remove rows with most of the columns that nan
for file_name in input_files_name:
    table = input_data[file_name]
    rows_to_be_removed = []
    for index, row in table.iterrows():
        if (row.isna().sum()/len(row)) >= 0.5:
            rows_to_be_removed.append(index)
    print(file_name, ' shape: ', table.shape)
    print('number of rows removed: ', len(rows_to_be_removed))
    input_data[file_name] = table.drop(rows_to_be_removed, axis=0)
    
        
```

```python
#     Replace the remaining null with median for number type and 'none category' for string type
for file_name in input_files_name:    
    table = input_data[file_name]
    number_columns = table.select_dtypes(include=['int64', 'float64'])
    string_columns = table.select_dtypes(include=['object'])
    for column_name in number_columns:
        column_mean = table[column_name].mean
        table[column_name].fillna(column_mean, inplace=True)
    for column_name in string_columns:
        table[column_name].fillna('none category', inplace=True)
    input_data[file_name] = table
```

```python
for file_name in input_files_name:
    df = input_data[file_name]
    # for column in df.columns:
    #    df[column] = df[column].map(lambda x: x.encode('unicode-escape').decode('utf-8'))
    df.to_pickle('../data/preprocessed/removed_nan_v2/' + file_name + '.pkl')
```
